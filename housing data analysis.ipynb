# scaling and binning data 

```python
import pandas as pd 
df = pd.read_csv('housing.csv')
df.head()
```

**Min-Max Scaling**

```python
df["house_value_minmax"] = (df.medilue-df.median_house_value.min()) / (df.median_house_value.max()-df.median_house_value.min())
df.headan_house_va
```

Z-Score

```python
df["house_value_z"] = (df.median_house_value - df.median_house_value.mean()) / df.median_house_value.std()
df.head()
```

Binnig 

```python
bin_names = ["cheap", "medium", "high", "luxury"]
df["price_range"] = pd.cut(df.house_value_minmax, 
                           [0, .25, .5, .75, 1],
                           labels=bin_names)
df.head()
```

```python
df["price_range_quantile"] = pd.qcut(df.house_value_minmax, 
                                     4, labels = bin_names)
df.head()
```

```python
df.price_range_quantile.unique()
```
